```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import os
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
dfWeather=pd.read_csv("WeatherDF.csv")
dfWeather = dfWeather.set_index('fullDate')
```

```python
dfWeather.head()
```

```python
dfWeather.describe()
```

```python
dfWeather.columns
```

```python
dfWeather.loc[dfWeather['season']=='winter','season']=1
dfWeather.loc[dfWeather['season']=='summer','season']=2
dfWeather.loc[dfWeather['season']=='fall','season']=3
dfWeather.loc[dfWeather['season']=='spring','season']=4
```

```python
dfWeather['season']=dfWeather['season'].astype(int)
```

```python
dfWeather.info()
```

```python
dfWeather_num=dfWeather[list(dfWeather.dtypes[dfWeather.dtypes!='object'].index)]
```

```python
dfWeather_num.pop('feelslike_c')
dfWeather_num
```

```python
y = dfWeather_num.pop('temp_c')
X = dfWeather_num
```

```python
dfWeather_num
```

# Spliting the data for training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=4)
```

```python
X_train.head()
```

```python
X_test
```

# Building the model

```python
model = LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
#calculate the error
np.mean((pred-y_test)**2)
```

```python
pd.DataFrame({'actual': y_test,
              'prediction': pred,
              'diff': (y_test-pred)})
```

```python
from sklearn import metrics, preprocessing
```

```python
print(f"mse: {metrics.mean_squared_error(y_test,pred)}")
print(f"r^2: {metrics.r2_score(y_test,pred)}")
```
